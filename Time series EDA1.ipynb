# Time Resampling

```python
#show all min values in year end frequency
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='C').min()
```

```python
df_tesla.resample(rule='A').max()['High'].plot()
```

```python
#Plot the Quartely frequency data
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
#Business end frequency
df_tesla.resample(rule='BA').max()['High'].plot()
```

```python
#plotting
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
#Get monthly avg data
df_tesla['Close'].resample(rule='M').mean()
```

```python
#Rolling
df_tesla['Open'] = df_tesla['Open'].rolling(10).mean()
```

```python
df_tesla['Open:30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla[['Open','Open:30 days rolling']].plot()
```

```python
#smoothening the curve
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,6))
```
